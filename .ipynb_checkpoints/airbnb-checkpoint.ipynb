# Clean data

```python
data = pd.read_csv('train.csv')
data.head() 
```

```python
data.shape
```

```python
data.columns.values
```

```python
data = data.drop(['thumbnail_url','review_scores_rating','name','neighbourhood','zipcode','amenities','host_response_rate','description', 'first_review',
       'host_has_profile_pic', 'host_identity_verified',
       'host_response_rate', 'host_since', 'instant_bookable',
       'last_review', 'latitude', 'longitude','cancellation_policy','bathrooms'], axis=1)
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
from IPython.display import clear_output
n = len(data)

for index, row in data.iterrows():
    
    if index==0:
        display('Processing: 0%')
    elif index%100 == 0:
        clear_output()
        display('Processing: ')
        display(str(round(index/n,2)*100)+" %")
        
    if row['property_type'] in data.columns.values:
        data.loc[index,'property_type'] = 1
    else:
        data[row['property_type']] = 0
        data.loc[index,row['property_type']] = 1
        
    if row['city'] in data.columns.values:
        data.loc[index,row['city']] = 1
    else:
        data[row['city']] = 0
        data.loc[index,row['city']] = 1 
                 
    if row['bed_type'] in data.columns.values:
        data.loc[index,row['bed_type']] = 1
    else:
        data[row['bed_type']] = 0
        data.loc[index,row['bed_type']] = 1
                 
    if row['cleaning_fee'] == "TRUE":    
        data.loc[index,'cleaning_fee'] = 1
    else:
        data.loc[index,'cleaning_fee'] = 0

data.head()
```

```python
data.to_csv("cleaned_airbnb.csv",sep=",")
```
